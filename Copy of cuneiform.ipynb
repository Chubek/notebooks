```
def estimate_shannon_entropy(dna_sequence):
    bases = collections.Counter([tmp_base for tmp_base in dna_sequence])
    # define distribution
    dist = [x/sum(bases.values()) for x in bases.values()]
 
    # use scipy to calculate entropy
    entropy_value = entropy(dist, base=2)
 
    return entropy_value
```

```
 ls_prob = []
for char, occ in dict(freq).items():
  if char == " ":
    continue
  
  ls_prob.append((char, occ, occ / len([d for d in dna if d != " "])))

df_prob = pd.DataFrame.from_records(ls_prob, columns=["character", "occurance", "frequency"])

df_prob
```

```
transmat = []

shuff = shuffle(df_prob["frequency"])

for i in df_prob["frequency"].values:
  sub_mat = []
  for j in shuff.values:
    sub_mat.append(i * j)

  transmat.append(sub_mat)


transmat
  
```

```
startprob = df_prob["frequency"].values

model = hmm.GaussianHMM(3, "full", startprob, transmat)
```
