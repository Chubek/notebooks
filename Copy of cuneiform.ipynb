```
def estimate_shannon_entropy(dna_sequence):
    bases = collections.Counter([tmp_base for tmp_base in dna_sequence])
    # define distribution
    dist = [x/sum(bases.values()) for x in bases.values()]
 
    # use scipy to calculate entropy
    entropy_value = entropy(dist, base=2)
 
    return entropy_value
```

```
 ls_prob = []
for char, occ in dict(freq).items():
  if char == " ":
    continue
  
  ls_prob.append((char, occ, occ / len([d for d in dna if d != " "])))

df_prob = pd.DataFrame.from_records(ls_prob, columns=["character", "occurance", "frequency"])

df_prob
```

```
transmat = []

shuff = shuffle(df_prob["frequency"])

for i in df_prob["frequency"].values:
  sub_mat = []
  for j in shuff.values:
    sub_mat.append(i * j)

  transmat.append(sub_mat)


transmat
  
```

```
startprob = df_prob["frequency"].values

model = hmm.GaussianHMM(3, "full", startprob, transmat)
```

```
remodel = hmm.GaussianHMM(n_components=3, covariance_type="full", n_iter=100)
remodel.fit(df_prob["frequency"].values.reshape(-1, 1))
```

```
remodel.predict(df_prob["frequency"].values.reshape(-1, 1))
```

```
ls_probs = []

for word in splt:
  count_dict = dict(collections.Counter(word))
  for char in word:
    ls_probs.append((char, word, count_dict[char] / len(word)))

  
df_probs = pd.DataFrame.from_records(ls_probs, columns=["character", "word", "probablity"])

df_probs

  
```

```
remodel = hmm.GaussianHMM(n_components=3, covariance_type="full", n_iter=100)
remodel.fit(df_probs["probablity"].values.reshape(-1, 1))
```

```
df_mult = df_probs.loc[:, ["character", "probablity"]].groupby("character").mean()

ls_obs = []

for ob in df_mult["probablity"].values:
  ob_row = []
  for ob_inner in df_mult["probablity"].values:
    ob_row.append(ob * ob_inner)

  ls_obs.append(ob_row)

ls_obs
```
